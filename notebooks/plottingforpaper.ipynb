## Plotting for paper

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

## Offshore CAWCR

```python
offStats = pd.read_csv("../data/stats_offshore.csv", index_col = 0 , parse_dates =True )
offStats = offStats.apply(pd.to_numeric, errors='coerce')
offStats = offStats.drop(['lat', 'lon'], axis=1)
```

```python
fig = plt.figure()
_ = offStats.plot(subplots = True, figsize = (10,10))
plt.savefig("offshore_modelled.eps")
```

## Target Data (SWAN)

```python
GCStats = pd.read_csv("../data/stats_GC.csv", index_col = 0 , parse_dates =True )
GCStats = GCStats.apply(pd.to_numeric, errors='coerce')
GCStats = GCStats.drop(['site'], axis=1)
```

```python
fig = plt.figure()
_ = GCStats.plot(subplots = True, figsize = (10,10))
plt.tight_layout()
plt.suptitle('SWAN Gold Coast Wave Monitoring Site')
plt.savefig('SWAN_GC.png')
```
