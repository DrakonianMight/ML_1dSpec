```python
import wavespectra
import matplotlib.pyplot as plt
```

```python
ds = wavespectra.read_swan('../data/CAWCR_201809_786.sp2')
```

```python
ds.isel(time =1, lat = 0, lon = 0).spec.oned().plot()
plt.title('')
plt.ylabel('Wave Spectral Density (m2/hz)')
plt.xlabel('Wave frequency (hz)')
plt.tight_layout()
plt.savefig('./plots/1dSpectrum.png')
```
